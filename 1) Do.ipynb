```python
import pandas as pd
import requests
import json
from PyDictionary import PyDictionary
dictionary=PyDictionary()
```

```python
words = pd.read_csv('words.csv')
```

```python
cols = [col for col in words.columns if "Group " in col]
```

```python
words = words[cols]
```

```python
words = words[1:]
```

```python
import functools
import operator
all_words = functools.reduce(operator.iconcat, words.values, [])
```

```python
word_list = [x for x in all_words if " " not in str(x) and not pd.isnull(x)]
```

```python
word_meaning = {}
```

```python
i = 1

for word in word_list:
    print("{} of {}".format(i, len(word_list)))
    if word not in word_meaning:
        word_meaning[word] = dictionary.meaning(word)
    i = i + 1
```

```python
word_meaning
```

```python
all_details = pd.DataFrame()
try:
    for word, meaning in word_meaning.items():
        curr_details = {}
        meaning_str = ""
        for word_type, vals in meaning.items():
            meaning_str = meaning_str + word_type + ":<br/>"
            for i, val in enumerate(vals):
                meaning_str = meaning_str + str(i+1) + ":" + val + "<br/>"
        
        curr_details['word'] = word
        curr_details['meaning'] = meaning_str
        all_details = all_details.append(curr_details, ignore_index=True)
except:
    pass
```

```python
all_details[['word', 'meaning']].to_csv('word_meaning.csv', index=None)
```
